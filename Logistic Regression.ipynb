```python
n = 15
i = 0 
j = n-1
r = n//2
q = n//2
for s in range(n):
    for d in range(n):
        if(s==0 or s==n-1 or d==0 or d==n-1 or d==n//2 or s==n//2):
            print("*",end=" ")
        elif(d==i or d==j or d==r or d==q):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    i += 1
    j -= 1
    if(s<n//2):
        r -= 1
        q += 1
    else:
        r += 1
        q -= 1
    print()

```

```python
n = 20
s = 0
e = n-1
for i in range(n):
    if(i==0 or i==n-1 or i==n//2 or i==n//2-1 or i==n//2+1):
        print("x",end=" ");
    else:
        print("*",end=" ");
print()
for i in range(n):
    if(i!=n//2):
        print("*",end=" ");
    else:
        print("-",end=" ");
print()
for i in range(n):
    print("*",end=" ");
print()
for i in range(n//2+1):
    for j in range(n):
        if(i==0):
            print("*",end=" ");
        elif(j>=s and j<=e):
            print("*",end=" ");
        else:
            print("-",end=" ");
    if(i < n//2):
        s += 1
        e -= 1
    print()
```
