```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
#Data Gathering
df=pd.read_csv('medical_insurance.csv') 
df
```

```python
df.info()
```

```python
df['sex'].value_counts()
```

```python
df['smoker'].value_counts()
```

```python
df['smoker'].replace({'yes':0,'no':1},inplace=True)
# df
# df['smoker'].replace({'':'yes',1:'no'},inplace=True)
# df
df['sex'].replace({1:'male',0:'female'},inplace=True)
df
```

```python
df['region'].value_counts()
```

```python
df1=pd.get_dummies(df,columns=['region'])
df1.info()
```

```python
df1['region_northeast'].replace({True:1,False:0},inplace=True)
df1['region_northwest'].replace({True:1,False:0},inplace=True)
df1['region_southeast'].replace({True:1,False:0},inplace=True)
df1['region_southwest'].replace({True:1,False:0},inplace=True)
df1
# df1.info()
```

```python
#Train test split
x=df1.drop(['sex','charges'],axis=1)
y=df1['charges']
```

```python
df1
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Model Training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
df1
```

```python
y_pred=linear_model.predict(x_test)
mse=mean_absolute_error(y_pred,y_test)
print("mse value",mse)

rmse=np.sqrt(mse)
print("rmse",rmse)

r2=r2_score(y_pred,y_test)
print("r2 value",r2)
```
